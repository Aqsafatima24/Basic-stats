# Standardization of numerical variables


## Standardization(z-score normalizatio)
 Standardization (or Z-score normalization) is a way to make numbers easier to compare by putting them on the same level. It works by taking each number, subtracting the average, and then dividing by how much the numbers differ from each other. After doing this, the numbers will be centered around 0. This helps when you need to compare different types of data fairly.

# Before standardization

```python
def get_stand(x):
    return ((x-x.mean())/x.std())
```

```python
x=numeric_df
get_stand(x)
```

```python
"Summary Statistics Before Standardization:"
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
numerical_columns
```

```python
scaler = StandardScaler()
scaler
```

```python
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
df[numerical_columns] 
```

# After standardization

```python
"Summary Statistics after Standardization:"
df.describe()
```

```python
for column in numerical_columns:
    plt.figure(figsize=(11, 4))
    plt.subplot(1, 2, 1)
    plt.hist(df[column], bins=17, color='pink', edgecolor='black')
    plt.title(f'Distribution of {column} (Before Standardization)')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.subplot(1, 2, 2)
    plt.hist(df[column] , bins=17, color='purple', edgecolor='black')
    plt.title(f'Distribution of {column} (After Standardization)')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

# Conversion of Categorical Data into Dummy Variables

### Need for Converting Categorical Data into Dummy Variables (One-Hot Encoding)
In many machine learning algorithms, you can't directly use categories as inputs. One-hot encoding is a way to turn these categories into numbers by creating separate columns for each category. Each column will have a 1 if the category is present and a 0 if it's not. This makes the categorical data easy for machine learning models to understand and work with.

```python
" Dataset Before One-Hot Encoding:"
df.head()
```

```python
df_encoded = pd.get_dummies(df,dtype=int)
```

```python
" Dataset After One-Hot Encoding:"
df_encoded.head()
```
